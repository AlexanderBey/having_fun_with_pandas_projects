```python
# Library: pandas, numpy, matplotlib, seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Importing the dataset
exchange_rates = pd.read_csv('./data-set/data.csv')

# Renaming columns
rename = exchange_rates.rename(columns={'[US dollar ]':'US_dollar',
                                        'Period\\Unit:':'Time'},
                               inplace=True)

# Converting Time column to datetime
convert_time = exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# Sorting values by Time in ascending order
sorting_time = exchange_rates.sort_values('Time', inplace=True)

# We reset the index and drop the initial index
reset_index = exchange_rates.reset_index(drop=True, inplace=True) 

# Isolate the columns Time and US_dollar
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]


# Control flow
rename
convert_time
sorting_time
reset_index


# Drop the rows where the - symbol appears
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

# Convert the US_dollar column to a float data type
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

# Showing the results
exchange_rates

euro_to_dollar

```

```python
# Plotting the line graph

import matplotlib.pyplot as plt
%matplotlib inline 
# Enables Jupyter to display graphs

plt.plot(euro_to_dollar['Time'], 
         euro_to_dollar['US_dollar'])
plt.show()
```
